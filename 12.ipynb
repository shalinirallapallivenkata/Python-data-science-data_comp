```python
season = pd.get_dummies(dat['season'], prefix = 'season')
weather = pd.get_dummies(dat['weather'], prefix = 'weather')
```

```python
# Concat all dummies and continuous variables to create predictors X
X = pd.concat([season,weather,
              dat[['hour','holiday','workingday',
                  'temp','atemp','humidity','windspeed', 'count_new']]
              ],
              axis=1)
X.info()
```

```python
# Normalize columns in X to range [0,1]
X_scale = preprocessing.minmax_scale(X, feature_range=(0, 1), axis=0)
X_scale = pd.DataFrame(X_scale)
X_scale.columns = X.columns
```

## 2.	Create design matrices to represent X and y.

```python
y, X = dmatrices('count_new ~ hour + season+ holiday + workingday + weather + temp + atemp + humidity + windspeed',
                 data=X,
                 return_type='dataframe')
```

```python
knn7 = neighbors.KNeighborsClassifier(n_neighbors = 7, 
                                      weights='uniform',                                    
                                      algorithm='auto')
knn7.fit(train_X, train_y)
pred_y_k7 = knn7.predict(test_X)
```
